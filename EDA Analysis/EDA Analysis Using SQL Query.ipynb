# order volume vary across different months 

```python
query = """
    SELECT MONTHNAME(order_purchase_timestamp) AS Month,
           Sum(product_photos_qty) AS order_volume
    FROM orders AS o
    JOIN order_items AS ol ON o.order_id = ol.order_id
    JOIN products AS p ON p.product_id = ol.product_id
    WHERE order_status = 'delivered'
    GROUP BY Month
    ORDER BY order_volume DESC
"""
df = pd.read_sql_query(query, connection)
plt.bar(df['Month'], df['order_volume'], color="purple")
plt.xlabel('Month')
plt.ylabel('Order Volume')
plt.title('Order Volume per Month')
plt.xticks(rotation=45)
plt.tight_layout() 
plt.show()
```

# relationships between the product category and the customer's location or state? Are certain categories more popular in specific regions

```python
query="""
      select customer_city,category,total_order from
(select customer_city, product_category_name_english as category ,count(p.product_id) as total_order,
dense_rank() over (partition by customer_city order by count(oi.product_id) desc) as rnk  from customers as c 
join orders as o on c.customer_id=o.customer_id 
join order_items as oi on o.order_id=oi.order_id
join products as p on oi.product_id=p.product_id
join product_category_name_translation as pc on pc.product_category_name=p.product_category_name
group by customer_city,category) as k
where rnk=1
order by total_order desc
limit 30

                         """

df = pd.read_sql_query(query, connection)
df

sns.barplot(data=df,x='customer_city',y='total_order',hue = 'category')
plt.xlabel('City')
plt.ylabel('Total Orders')
plt.title('Top Product Categories per City')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout() 
plt.show()
```

# particular payment methods that are more commonly used by customers

```python
query="""
      select payment_type,count(c.customer_id) as total_customers from order_payments as op
join orders as o on op.order_id=o.order_id
join customers as c on c.customer_id=o.customer_id
group by payment_type
order by total_customers desc
limit 4 
       """

df=pd.read_sql_query(query,connection)
df

labels = df['payment_type']
sizes = df['total_customers']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Payments type according to customer usages')
plt.tight_layout()
plt.show()
```

# Sates where seller generate the most revenue

```python
query="""
      select seller_state ,sum(price) as Revenue from sellers as s 
join order_items as oi on s.seller_id=oi.seller_id
group by seller_state
order by Revenue desc
limit 10

                    """

df= pd.read_sql_query(query,connection)
df

plt.bar(df['seller_state'] , df['Revenue'] ,color = 'blue')

plt.xlabel('Seller State')
plt.ylabel('Revenue')
plt.title('Revenue by Seller State')
plt.xticks(rotation=45)
plt.tight_layout() 
plt.show()
```

#  Avg delivery time vary based on different factors such as the seller, product category, or customer location

```python
query="""
        select  o.order_id,customer_state,product_category_name_english as category,order_status,
round(avg(datediff(order_delivered_customer_date,order_purchase_timestamp)),0) as Avg_delivery_time
from customers as c join orders as o on c.customer_id=o.customer_id
join order_items as oi on o.order_id=oi.order_id
join products as p on p.product_id=oi.product_id
join product_category_name_translation as pp on p.product_category_name=pp.product_category_name
group by order_id,customer_state,product_category_name_english,order_status
                                                                               """

df=pd.read_sql_query(query,connection)
df
```

# states that consistently have a higher average order value or purchase frequency

```python
query="""
       select customer_state as state ,count(o.order_id) as total_orders ,
round(avg(price+freight_value),2) as avg_order_value 
from customers as c 
join orders as o on c.customer_id=o.customer_id
join order_items as oi on o.order_id=oi.order_id
group by state
order by  avg_order_value  desc 
limit 10
        """

df=pd.read_sql_query(query,connection)
df
```

```python
plt.figure(figsize=(10, 6))
bars = plt.bar(df['state'], df['avg_order_value'], color='blue')
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.1f}', ha='center', va='bottom')
plt.xlabel('State')
plt.ylabel('Average Order Value')
plt.title('Top 10 States with Highest Average Order Value')
plt.xticks(rotation=45)
plt.tight_layout()  
plt.show()
```

# order status (e.g., delivered, shipped, canceled) vary across different product categories or sellers

```python
query="""
       select product_category_name_english as category,order_status,
count(*) as Total_orders from orders as o 
join order_items as oi on o.order_id=oi.order_id
join products as p on p.product_id=oi.product_id
join product_category_name_translation as pp on p.product_category_name=pp.product_category_name
group by category,order_status
order by count(*) desc 

                       """

df=pd.read_sql_query(query,connection)
df
```

# sales trend line in different month and year

```python
query="""
       select monthname(order_purchase_timestamp) as Month ,year(order_purchase_timestamp) as Year ,sum(price+freight_value) as revenue
from orders as o join order_items as oi on o.order_id=oi.order_id
group by Month,year
order by month,year
                   """

df=pd.read_sql_query(query,connection)
df


plt.figure(figsize=(10, 4))
plt.plot(df['Month'] + ' ' + df['Year'].astype(str), df['revenue'], marker='o', linestyle='-', color='green')
plt.xlabel('Month and Year')
plt.ylabel('Revenue')
plt.title('Monthly and Yearly Sales Revenue Trend')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()






```
